```python
#get the percentage of students who passed math and reading, as well as the overall passing percentage.
#first determine the passing grade by writing the code that does the following:
#To get the percentage of students who passed math and reading, we will write code to:
##1-Determine the passing grade.
##2-Get the number of students who passed math and reading in separate DataFrames.
##3-Calculate the number of students who passed math and reading.
##4-Calculate the percentage of students who passed math and reading.

### 1- first we are determining the records that passed.


#get the average math and reading scores
#continue with math average scores
#Add the Pandas dependency - STUDENT DATA & SCHOOL DATA
import pandas as pd

#Files to load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#Read the student data file and store it in a Pandas DataFrame
student_data_df = pd.read_csv(students_data_to_load)
student_data_df.count()

#Read the school data file and store it in a Pandas DataFrame
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head()

# Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()

# Get the total number of students.
student_count = school_data_complete_df.count()
student_count

#school_data_complete_df["Student ID"].count()
#To assign the "student_count" to a column that identifies with students, we will use the "Student ID" column.

student_count = school_data_complete_df["Student ID"].count()
student_count

print(student_count, "student count")

#There are two ways
# 1 -Calculate the total number of schools.
# school_count = school_data_df["school_name"].count()
# school_count
# 2 - Calculate the total number of schools using unique and provide count instead of list
school_count_2 = len(school_data_complete_df["school_name"].unique())
school_count_2
print(school_count_2, "school count")
#len adds the count to the results. without it we get a list of high schools

#to calculate the total budget we add all the fo them from a list. remember to use brackets
total_budget = school_data_df["budget"].sum()
total_budget
print(total_budget, "Total Budget")
# Calculate the average reading score.
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
print (average_reading_score, "average reading score")

# Calculate the average math score.
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
print(average_math_score, "average math score")

###1-Determine the passing grade.

####passing score was 70 for math and reading
####math and reading scores are lists

passing_math = school_data_complete_df["math_score"] >=70
passing_math
passing_reading = school_data_complete_df["reading_score"] >= 70
passing_reading

##2-Get the number of students who passed math and reading in separate
#MATH
# Get all the students who are passing math in a new DataFrame by filtering
#the dataframe for the word true by adding our above code in brackets
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()

##READING
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >=70]
passing_reading.head()

##3 get the number of students who passed each subject
##MATH
passing_math_count = passing_math["student_name"].count()
##READING
passing_reading_count = passing_reading["student_name"].count()

print(passing_math_count,"passed math")
print(passing_reading_count, "passed reading")

##4-Calculate the percentage of students who passed math and reading.

#Calculate the percent that passed math.
passing_math_percentage = passing_math_count / float(student_count) * 100

# Calculate the percent that passed reading.
passing_reading_percentage = passing_reading_count / float(student_count) * 100

print(passing_math_percentage, "math passed percentage")
print(passing_reading_percentage,"reading passed percentage")

#Calculate the overall passing number for both subjects
##first find from the list of merged records which students passed for math and reading
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
passing_math_reading.head()

##then find the total number of students that passed both subjects
overall_passing_math_reading_count = passing_math_reading["student_name"].count()

print(overall_passing_math_reading_count, "Students that passed both math and reading")


## finally calculate the percentage of students that passed both
overall_passing_percentage = overall_passing_math_reading_count/student_count * 100
overall_passing_percentage

print(overall_passing_percentage,"Passing percentages for students that passed both subjects")
```
